```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 载入 .mat 数据
mat = scipy.io.loadmat(r'D:\CLS\Thesis\Qi_CODE\PHI014\Preprocessed_PHI014_FSit.mat')

# 提取数据
time = mat['data']['time'][0][0].squeeze()
time2 = time[::2]
tsi = mat['data']['TSI'][0][0]
tsi_ff = mat['data']['TSI_FF'][0][0]
```

```python
# 绘图：TSI(Tissue Saturation Index)
plt.figure(figsize=(10, 5))
plt.plot(time2, tsi[:, 1], 'deepskyblue', linewidth=0.5)
plt.xlabel('Time(s)')
plt.ylabel('TSI(%)')
plt.legend(['TSI(Tissue Saturation Index)'])
plt.title('TSI Changes with Time')
plt.grid(True)
plt.show()
```

```python
t_start1 = 0
t_end1 = 485
mask1 = (time2 >= t_start1) & (time2 <= t_end1)
time2_1 = time2[mask1]
tsi_1 = tsi[mask1, 1]

plt.figure(figsize=(12, 5))
plt.plot(time2_1, tsi_1, color='deepskyblue', linewidth=0.5)
plt.xlabel('Time (s)')
plt.ylabel('TSI (%)')
plt.title('TSI Changes with Time')
plt.legend(['TSI (Tissue Saturation Index)'])

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据并压缩为一维
t_eval_trans = pd.read_csv('t_eval_trans.csv', delimiter=',').squeeze()
store_SaO2 = pd.read_csv('store_SaO2.csv', delimiter=',').squeeze()

print("t_eval_trans shape:", t_eval_trans.shape)
print("store_SaO2 shape:", store_SaO2.shape)

# 确保两个数组长度相同
min_length = min(len(t_eval_trans), len(store_SaO2))
t_eval_trans = t_eval_trans[:min_length]
store_SaO2 = store_SaO2[:min_length]

# 设置绘图时间窗口
t_start = -100
t_end = 100
mask = (t_eval_trans >= t_start) & (t_eval_trans <= t_end)

t = t_eval_trans[mask]+100
sao2 = store_SaO2[mask]

# 绘图
plt.figure(figsize=(8, 5), dpi=300)
plt.plot(t, sao2*100, color='red', label='SaO2', linewidth=0.5)

plt.title('Oxygen Saturation CVSIM', fontsize=14)
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('SaO2 (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
