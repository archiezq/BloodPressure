```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Load .mat file
data = sio.loadmat('Preprocessed_PHI014_FSit.mat')

# Extract data
time = data['data']['time'][0, 0].flatten()
right = data['data']['cbfv_r'][0, 0].flatten()
left = data['data']['cbfv_l'][0, 0].flatten()
pressure = data['data']['BP'][0, 0].flatten()
tsi = data['data']['TSI'][0, 0]
tsiff = data['data']['TSI_FF'][0, 0]
time10 = data['data']['time_10Hz'][0, 0].flatten()
absO2Hb = data['data']['absO2Hb'][0, 0].flatten()
absHHb = data['data']['absHHb'][0, 0].flatten()
```

```python
# Plot CBFV Right vs Left
plt.figure(figsize=(8,6))
plt.plot(time, right, 'r', linewidth=1.5, label='Right')
plt.plot(time, left, 'b', linewidth=1.5, label='Left')
plt.xlabel('Time (s)')
plt.ylabel('velocity (cm/s)')
plt.legend()
plt.title('CBFV Right vs. Left')
plt.grid(True)
plt.show()


# TSI Changes with Time
time2 = time[::2]
plt.figure(figsize=(8,6))
plt.plot(time2, tsi[:, 0], 'r', linewidth=1.5, label='TSI 1')
plt.plot(time2, tsi[:, 1], 'b', linewidth=1.5, label='TSI 2')
plt.xlabel('Time (s)')
plt.ylabel('TSI (%)')
plt.legend()
plt.title('TSI Changes with Time')
plt.grid(True)
plt.show()

# Tsiff Changes with Time
plt.figure(figsize=(8,6))
plt.plot(time2, tsiff[:, 0], 'r', linewidth=1.5, label='TSIFF 1')
plt.plot(time2, tsiff[:, 1], 'b', linewidth=1.5, label='TSIFF 2')
plt.xlabel('Time (s)')
plt.ylabel('TSIFF (%)')
plt.legend()
plt.title('TSIFF Changes with Time')
plt.grid(True)
plt.show()

# O2Hb, HHb, and TSI Changes with Time
plt.figure(figsize=(8,6))
plt.plot(time2, absO2Hb, 'r', linewidth=1.5, label='O2Hb')
plt.plot(time2, absHHb, 'b', linewidth=1.5, label='HHb')
plt.plot(time2, absHHb + absO2Hb, 'k', linewidth=1.5, label='Total Hb')
# ttsi = absO2Hb / (absHHb + absO2Hb) * 100
# plt.plot(time2, ttsi, 'g', linewidth=1.5, label='TSI')
# plt.plot(time2, tsi[:, 1], 'red', linewidth=1.0, label='TSI 2')
plt.xlabel('Time(s)')
plt.ylabel('Concentration (umol/L)')
plt.legend()
plt.title('O2Hb and HHb Changes with Time')
plt.grid(True)
plt.show()

# O2Hb, HHb, and TSI Changes with Time
plt.figure(figsize=(8,6))
plt.plot(time2, absO2Hb, 'r', linewidth=1.5, label='O2Hb')
plt.xlabel('Time(s)')
plt.ylabel('Concentration (umol/L)')
plt.legend()
plt.title('O2Hb Changes with Time')
plt.grid(True)
plt.show()

# O2Hb, HHb, and TSI Changes with Time
plt.figure(figsize=(8,6))
plt.plot(time2, absHHb, 'b', linewidth=1.5, label='HHb')
plt.xlabel('Time(s)')
plt.ylabel('Concentration (umol/L)')
plt.legend()
plt.title('HHb Changes with Time')
plt.grid(True)
plt.show()


```
